# 1. Vector

A Vector has two attributes, magnitude and direction. 


## 1.1 Two-dimensional Vector
A two-dimensional vector (or 2-D vector) means there are two numbers (or two items) in a vector, like above examples.
Below code snippet defines the victors in python, and visualize them in a 2D space.

Example of a vector:

$\vec{v} = \begin{bmatrix}2\\3\end{bmatrix}$, or $\begin{bmatrix}2&3\end{bmatrix}$

It doesnt't matter if it is in a vertical column, or in a horizontal row.

The magnitude of a vector is calculated by:

$\left | \vec{v} \right | = \sqrt{x^2 + y^2} = \sqrt{2^2 + 3^2}$



```python
v = np.array([2,3])
origin = np.array([0, 0]) # origin point

# Creating plot
fig = plt.figure(figsize=(6,6))
plt.title('2-D Vector')
plt.quiver(*origin, *v, color='k', units='xy', scale=1)
plt.text(*v, r'$\vec{v} = [2,3]$', fontsize=20, color='k')

# x-lim and y-lim
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.arrow(-1, 0, 4.75, 0, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.arrow(0, -1, 0, 4.75, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.text(0.1, 3.75, r"$y$", fontsize=16)
plt.text(3.75, -0.3, r'$x$', fontsize=16)

plt.plot([2,v[0]],[0,v[1]], color='k',linestyle='dashed')
plt.plot([0,v[0]],[3,v[1]], color='k',linestyle='dashed')
plt.text(*v+[0,-1.5], r'$3$', fontsize=20, color='k')
plt.text(*v+[-1,0.1], r'$2$', fontsize=20, color='k')
plt.text(*v+[-1.9,-1], r'$\sqrt{2^2+3^2}$', fontsize=20, color='k')

# Show plot with grid
plt.grid(which='major', color='#555555', linestyle='--')
plt.savefig("vector_2d_plot2.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```

## 1.2 Three or more dimensional Vector
A 3-dimensional vector (or 3-D vector) means there are three items in a vector, e.g.

$\vec{v} = \begin{bmatrix}3 & 2 & 2\end{bmatrix}$, 

$\vec{u} = \begin{bmatrix}1\\-3\\-1\end{bmatrix}$, 

$\vec{w} = \begin{bmatrix}-3 & 1 & 2\end{bmatrix}$

Below code snippet defines the three vectors, and visualize them in a 3-D space.

The magnitude is calculated by:

$\left | \vec{v} \right | = \sqrt{x^2 + y^2 + z^2} = \sqrt{3^2 + 2^2 + 3^2}$

```python
# draw vector
v = np.array([3,2,2])
u = np.array([1,-3,-1])
w = np.array([-3,1,2])

print("v = ", v, "; |v| =", np.linalg.norm(v) )
print("u = ", u, "; |u| =", np.linalg.norm(u) )
print("w = ", w, "; |w| =", np.linalg.norm(w) )

origin = np.array([0, 0, 0]) # origin point
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.quiver(*origin, *v, arrow_length_ratio=0.1, colors='k')
ax.text(*v, r'$\vec{v}$', fontsize=20, color='k')
ax.quiver(*origin, *u, arrow_length_ratio=0.2, colors='k')
ax.text(*u+[0,0.1,-0.8], r'$\vec{u}$', fontsize=20, color='k')
ax.quiver(*origin, *w, arrow_length_ratio=0.1, colors='k')
ax.text(*w, r'$\vec{w}$', fontsize=20, color='k')
ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])
ax.set_zlim([-4, 4])
ax.set_title("3-D Vectors")
plt.savefig("vector_3d_plot.svg", format="svg", dpi=150, transparent=True, bbox_inches='tight')
plt.show()
```

## 1.3 Vector Addition

Say there are two vectors:

$\vec{v} = \begin{bmatrix}2\\3\end{bmatrix},\space\space\vec{u} = \begin{bmatrix}-1\\2\end{bmatrix}$, 

The addition of the two vectors:

$\vec{w} = \vec{v} + \vec{u} = \begin{bmatrix}2\\3\end{bmatrix} + \begin{bmatrix}-1\\2\end{bmatrix} = \begin{bmatrix}2 + (-1)\\3 + 2\end{bmatrix} = \begin{bmatrix}1\\5\end{bmatrix} $

```python
v = np.array([2,3])
u = np.array([-1,2])
w = v + u
origin = np.array([0, 0]) # origin point

# Creating plot
fig = plt.figure(figsize=(6,6))
plt.title('2-D Vector Addition')

plt.quiver(*origin, *v, color='k', units='xy', scale=1)
plt.text(*v, r'$\vec{v}$', fontsize=20, color='k')
plt.quiver(*origin, *u, color='k', units='xy', scale=1)
plt.text(*u+[-0.3,0], r'$\vec{u}$', fontsize=20, color='k')
plt.quiver(*origin, *w, color='navy', units='xy', scale=1)
plt.text(*w, r'$\vec{w}=\vec{v}+\vec{u}$', fontsize=20, color='navy')

plt.plot([v[0],w[0]],[v[1],w[1]], color='navy',linestyle='dashed')
plt.plot([u[0],w[0]],[u[1],w[1]], color='navy',linestyle='dashed')

# x-lim and y-lim
plt.xlim(-3, 4)
plt.ylim(-1, 6)
plt.arrow(-3, 0, 6.75, 0, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.arrow(0, -1, 0, 6.75, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.text(0.1, 5.75, r"$y$", fontsize=16)
plt.text(3.75, -0.3, r'$x$', fontsize=16)

# Show plot with grid
plt.grid(which='major', color='#555555', linestyle='--')
# plt.grid(which='minor', color='#555555', linestyle=':')
plt.savefig("vector_2d_addition.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```

## 1.5	Vector Norm and Distance

**Euclidean distance:** the distance between two points which is the length of line segment between the two points.

There are two points $(x_1, y_1)$ and $(x_2, y_2)$, the euclidean distance is:

distance $= \sqrt{(x_1-y_1)^2 + (x_2-y_2)^2}$

**Vector Norm:** the distance of vector from the origin. It's also referred as magnitude or length of the vector.

There is a vector $\vec{v}  = \begin{bmatrix}v_1 & v_2 & ... & v_n\end{bmatrix}$, its norm is:

$\parallel \vec{v} \parallel = \sqrt{v_1^2 + v_2^2 + ... + v_n^2}$

**Vector Distance:** the distance between two vectors. It's the norm of the vector subtraction.

There are two vectors $\vec{v}  = \begin{bmatrix}v_1 & v_2 & ... & v_n\end{bmatrix}$, 

and $\vec{u}  = \begin{bmatrix}u_1 & u_2 & ... & u_n\end{bmatrix}$

The distance between the two is:

$\parallel \vec{v} - \vec{u} \parallel = \sqrt{(v_1-u_1)^2 + (v_2-u_2)^2 + ... + (v_n-u_n)^2}$

Vector norm:

```python
v = np.array([2,5,3,-1,3,7,5])
norm = np.linalg.norm(v)
print(norm)
```

```python
norm = np.sqrt(np.sum(v**2))
print(norm)
```

Vector distance:

```python
v = np.array([2,5,3,-1,3,7,5])
u = np.array([1,4,2,-4,-1,2,3])
distance = np.linalg.norm(v-u)
print(distance)
```

```python
distance = np.sqrt(np.sum((v-u)**2))
print(distance)
```

### 1.6 Scalar multiplication

A scalar is a number, to multiply a vector by a scalar means to multiply each item of the vector by the scalar. The scalar multiplication will change the magnitude (or the size) of the vector, but will not change the direction.

Here is the formula,

$\vec{w} = 2 \cdot\begin{bmatrix}2 & 3\end{bmatrix} = \begin{bmatrix}2 \cdot 2 & 2 \cdot 3\end{bmatrix} = \begin{bmatrix}4 & 6\end{bmatrix}$

In generic case:

$\vec{w} = \lambda\begin{bmatrix}v_1 & v_2 & ... & v_n\end{bmatrix} = \begin{bmatrix}\lambda v_1 & \lambda v_2 & ... & \lambda v_n\end{bmatrix}$



Below is an example of scalar multiplication for a 2-D vector by Python:

```python
v = np.array([2,3])
u = 2*v
print("v = ", v)
print("u = 3*v =", u)
```

```python
origin = np.array([0, 0]) # origin point

# Creating plot
fig = plt.figure(figsize=(6,6))
plt.title('2D Vector Scalar Multiplication')

plt.quiver(*origin, *u, color='navy', units='xy', scale=1)
plt.text(*u, r'$\vec{u}=2\vec{v}$', fontsize=20, color='navy')
plt.quiver(*origin, *v, color='k', units='xy', scale=1)
plt.text(*v+[-0.4,0], r'$\vec{v}$', fontsize=20, color='k')


# x-lim and y-lim
plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.arrow(-1, 0, 7.75, 0, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.arrow(0, -1, 0, 7.75, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.text(0.1, 6.65, r"$y$", fontsize=16)
plt.text(6.75, -0.3, r'$x$', fontsize=16)

# Show plot with grid
plt.grid(which='major', color='#555555', linestyle='--')
# plt.grid(which='minor', color='#555555', linestyle=':')
plt.savefig("vector_2d_scalar_mul.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```

## 1.7 Dot Product

There are two vectors:

$\vec{v}=\begin{bmatrix} v_1 & v_2 & ... & v_n \end{bmatrix}$

$\vec{u} = \begin{bmatrix}u_1\\u_2\\...\\u_n\end{bmatrix}$

There two representations of the dot product of vectors -- algebraic and geometric.

The algebraic representation of dot product for $\vec{v}$ and $\vec{u}$ is:

$\begin{array}{rcl}
\vec{v} \cdot \vec{u} & = & \begin{bmatrix}v_1 & v_2 & ... & v_n\end{bmatrix} \cdot 
\begin{bmatrix}u_1\\u_2\\...\\u_n\end{bmatrix} \\
 & = & v_1 u_1 + v_2 u_2 + ... + v_n u_n\\
 & = & \displaystyle\sum_{i=1}^{n} v_i u_i\\
\end{array}$



The sum or square can be expressed as dot product of same vector:

$\begin{array}{rcl}
\vec{v} \cdot \vec{v} & = & \begin{bmatrix}v_1 & v_2 & ... & v_n\end{bmatrix} \cdot 
\begin{bmatrix}v_1\\v_2\\...\\v_n\end{bmatrix} \\
 & = & v_1^2 + v_2^2 + ... + v_n^2\\
 & = & \displaystyle\sum_{i=1}^{n} v_i^2\\
\end{array}$


The result of dot product of two vectors is a number (or scalar), not a vector. In Python *numpy.dot(u, v)* is used to calculate the dot product.

```python
v = np.array([2,3])
u = np.array([-1,2])
w = np.dot(v, u)
print("Dot production of v and u =", w)
```

The geometric representation of dot product of two vectors is:

$\vec{v} \cdot \vec{u} = \parallel \vec{v} \parallel \cdot \parallel \vec{u} \parallel \cdot cos(\theta)$

$\theta$ is the angle between $\vec{v}$ and $\vec{u}$.

$\parallel \vec{v} \parallel $ is the magnitude of vector $\vec{v}$, and $\parallel \vec{u} \parallel $ is the magnitude of vector $\vec{u}$.

```python
from matplotlib.patches import Arc

origin = np.array([0, 0]) # origin point

# Creating plot
fig = plt.figure(figsize=(6,6))
plt.title('2-D Vector Dot Product')
ax = plt.gca()

plt.quiver(*origin, *v, color='k', units='xy', scale=1)
plt.text(*v, r'$\vec{v}$', fontsize=20, color='k')
plt.quiver(*origin, *u, color='navy', units='xy', scale=1)
plt.text(*u+[-0.3,0], r'$\vec{u}$', fontsize=20, color='navy')

# x-lim and y-lim
plt.xlim(-2, 3)
plt.ylim(-1, 4)
plt.arrow(-3, 0, 5.75, 0, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.arrow(0, -1, 0, 4.75, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.text(0.1, 3.75, r"$y$", fontsize=16)
plt.text(2.75, -0.3, r'$x$', fontsize=16)
pac = Arc([0, 0], 1, 1, angle=0, theta1=55, theta2=118)
ax.add_patch(pac)
plt.text(0, 0.6, r"$\theta$", fontsize=20)
plt.plot([u[0],0.6],[u[1],0.9], color='navy',linestyle='dashed')
plt.plot([origin[0],0.6],[origin[1],0.9], color='navy',linestyle='solid', linewidth=2)
plt.text(0.4, 0.3, r"$\parallel \vec{u} \parallel \cdot cos(\theta)$", color='navy', fontsize=14)
plt.text(0.85, 2.2, r"$\parallel \vec{v} \parallel $", color='k', fontsize=14)

plt.text(1.1, 1.4, r"$\vec{v} \cdot \vec{u} = $", color='k', fontsize=16)
plt.text(1.05, 1.15, r"$\parallel \vec{v}\parallel \cdot \parallel \vec{u} \parallel \cdot cos(\theta)$", color='k', fontsize=14)


# Show plot with grid
plt.grid(which='major', color='#555555', linestyle='--')
# plt.grid(which='minor', color='#555555', linestyle=':')
plt.savefig("vector_2d_dotProduct.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```

## 1.8 Cross Product

The cross product of $\vec{v}$ and $\vec{u}$, denoted as $\vec{v} \times \vec{u}$, is a vector that is perpendicular to both $\vec{v}$ and $\vec{u}$, the magnitude is equal to the area of the parallelogram shaped by both vectors.

$\vec{v} \times \vec{u} = \parallel \vec{v}\parallel\cdot \parallel\vec{u}\parallel\cdot sin(\theta)$

The direction of the cross product follows the [right-hand rule](https://en.wikipedia.org/wiki/Right-hand_rule).

In Python numpy.cross(u, v) is used to calculate the cross product.

```python
v = np.array([2,-1,0])
u = np.array([1,1,0])
w = np.cross(v, u)
print("w = v x u = ", w)

origin = np.array([0, 0, 0]) # origin point
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.quiver(*origin, *v, arrow_length_ratio=0.2, colors='k')
ax.text(*v, r'$\vec{v}$', fontsize=20, color='k')
ax.quiver(*origin, *u, arrow_length_ratio=0.2, colors='k')
ax.text(*u, r'$\vec{u}$', fontsize=20, color='k')
ax.quiver(*origin, *w, arrow_length_ratio=0.1, colors='navy')
ax.text(*w, r'$\vec{w}=\vec{v} \times \vec{u}$', fontsize=20, color='navy')
ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])
ax.set_zlim([0, 4])
ax.set_title("2-D Vector Cross Product")
plt.savefig("vector_2d_crossProduct.svg", format="svg", dpi=150, transparent=True, bbox_inches='tight')
plt.show()
```

## 1.9 Cosine Similarity

```python
v = np.array([2,3])
u = np.array([-1,2])
w = np.dot(v, u)

origin = np.array([0, 0]) # origin point

# Creating plot
fig = plt.figure(figsize=(6,6))
plt.title('Cosine Similarity')
ax = plt.gca()

plt.quiver(*origin, *v, color='k', units='xy', scale=1)
plt.text(*v, r'$\vec{v}$', fontsize=20, color='k')
plt.quiver(*origin, *u, color='k', units='xy', scale=1)
plt.text(*u+[-0.3,0], r'$\vec{u}$', fontsize=20, color='k')

# x-lim and y-lim
plt.xlim(-2, 3)
plt.ylim(-1, 4)
plt.arrow(-3, 0, 5.75, 0, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.arrow(0, -1, 0, 4.75, head_width=0.1, head_length=0.2, fc='k', ec='k')
plt.text(0.1, 3.75, r"$y$", fontsize=16)
plt.text(2.75, -0.3, r'$x$', fontsize=16)
pac = Arc([0, 0], 1, 1, angle=0, theta1=55, theta2=118)
ax.add_patch(pac)
plt.text(0, 0.6, r"$\theta$", fontsize=20)

# Show plot with grid
plt.grid(which='major', color='#555555', linestyle='--')
# plt.grid(which='minor', color='#555555', linestyle=':')
plt.savefig("cosine_similarity.svg", format="svg", dpi=300, bbox_inches='tight')
plt.show()
```

Cosine similarity is defined as:

$\quad\text{Cosine similarity} = \cos(\theta) = \displaystyle\frac{\vec{u}\cdot\vec{u}}{\parallel\vec{u}
\parallel\parallel\vec{v}\parallel}$

It can be calculated as:

```python
v = np.array([2,3])
u = np.array([-1,2])

cos_sim = np.dot(v, u) / \
          (np.linalg.norm(v)*np.linalg.norm(u))
print("Cosine Similarity is:", cos_sim)
```

Alternatively the cosine similarity matrix can be calculated by:

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim2 = cosine_similarity([v, u])
print("Cosine Similarity matrix:\n", cos_sim2)
```
